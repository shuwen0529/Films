```python
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
```

```python
movies = pd.read_csv("Resources/US_Movies_1972-2016.csv", encoding ="ISO-8859-1")
movies.drop(columns = ["Position", "Created", "Modified", "Description", "URL", "Title Type"], inplace=True) 
movies.head()
```

```python
# Retrieve data from OMDb by Const
api_key = "apikey=trilogy"
url = "http://www.omdbapi.com/?"+ api_key+"&i="

# Variables sorted by movies
Title = []
Rating_imdb = []
Rating_RT = []
Rating_Metacritic = []
BoxOffice = []
Runtime = []
Year = []
Num_Votes = []
Release_Date = []
Directors = []
Genres = []
Productions = []
Actors = []
Rates = []

# Variables sorted by counts
Actor = []
Actor_count = []
Genre = []
Genre_count = []
Production = []
Production_count = []
Rated = []
Rated_count = [] 
Director = []
Director_count = []
```

```python
count = 0 
for movie in movies["Const"]:
    try:
        response = requests.get(url + movie)
        movie_data = response.json()
        print("Processing "+str(count)+" : "+movie_data["Title"])
        Title.append( movie_data["Title"])
        ### Get RunTime
        Runtime.append( movie_data["Runtime"])
        ### Get Year
        Year.append( movie_data["Year"])
        ### Get Number Votes
        Num_Votes.append( movie_data["imdbVotes"])
        ### Get Release Date
        Release_Date.append( movie_data["Released"])
        ### Get RT Rating
        try:
            if movie_data["Ratings"][0]["Source"] =='Rotten Tomatoes':
                Rating_RT.append( movie_data["Ratings"][0]["Value"] )
            elif movie_data["Ratings"][1]["Source"] =='Rotten Tomatoes':
                Rating_RT.append( movie_data["Ratings"][1]["Value"] ) 
            elif movie_data["Ratings"][2]["Source"] =='Rotten Tomatoes':
                Rating_RT.append( movie_data["Ratings"][2]["Value"] ) 
            else:
                Rating_RT.append("N/A")
        except Exception:
            Rating_RT.append( "N/A" )
        ### Get imdb Rating
        Rating_imdb.append(movie_data["imdbRating"]) 
        ### Get Metacritic Rating
        Rating_Metacritic.append(movie_data["Metascore"]) 
        ### Get Box Office
        try:
            BoxOffice.append( movie_data["BoxOffice"] )
        except Exception:
            BoxOffice.append( "N/A" )
        ### Get Rates 
        try:
            Rates.append( movie_data["Rated"])
        except Exception:
            Rates.append( "N/A" )
        ### Get Directors
        try:
            Directors.append( movie_data["Director"])
        except Exception:
            Directors.append( "N/A" )
        ### Genre    
        try:
            Genres.append( movie_data["Genre"])
        except Exception:
            Genres.append( "N/A" )        
        ### Productions    
        try:
            Productions.append( movie_data["Production"])
        except Exception:
            Productions.append( "N/A" )
        ### Actors
        try:
            Actors.append( movie_data["Actors"])
        except Exception:
            Actors.append( "N/A" )
        ################################################################
        ### Get Actor Counts
        movie_Actors = movie_data["Actors"]
        Actor_names  = [x.strip() for x in movie_Actors.split(',')]
        for n in Actor_names:
            if n not in Actor:
                Actor.append(n)
                Actor_count.append(1)
            else:
                log_a = Actor.index(n)
                Actor_count[log_a] = Actor_count[log_a]+1
        ### Get Genre Counts
        movie_Genre = movie_data["Genre"]
        Genre_type  = [x.strip() for x in movie_Genre.split(',')]
        for g in Genre_type:
            if g not in Genre:
                Genre.append(g)
                Genre_count.append(1)
            else:
                loc_g = Genre.index(g)
                Genre_count[loc_g] = Genre_count[loc_g]+1
        ### Get Production
        try:
            if movie_data["Production"] not in Production:
                Production.append(movie_data["Production"])
                Production_count.append(1)
            else:
                loc_p = Production.index(movie_data["Production"])             
                Production_count[loc_p] =  Production_count[loc_p] +1 
        except Exception:
            print ("No Production for "+movie_data["Title"])
        ### Get Rated Counts
        if movie_data["Rated"] not in Rated:
            Rated.append(movie_data["Rated"])
            Rated_count.append(1)
        else:
            loc_r = Rated.index(movie_data["Rated"])
            Rated_count[loc_r] =  Rated_count[loc_r] +1
        ### Get Director Counts
        if movie_data["Director"] not in Director:
            Director.append(movie_data["Director"])
            Director_count.append(1)
        else:
            loc_d = Director.index(movie_data["Director"])
            Director_count[loc_d] =  Director_count[loc_d] +1
        count = count+1
    except Exception:
        print("Can't find "+movie+". Skipping")  #  movies.drop(count, axis=0, inplace=True) 
       
print("-------- Done with processing -----------")
```

```python
Actors_df = pd.DataFrame({"Actor Names": Actor, "Actor Counts": Actor_count})
Actors_df.to_csv("Resources/US_movies_Actors.csv", encoding='utf-8', index=True)

Genre_df = pd.DataFrame({"Genre": Genre, "Genre Counts": Genre_count})
Genre_df.to_csv("Resources/US_movies_Genre.csv", encoding='utf-8', index=True)

Production_df = pd.DataFrame({"Production": Production, "Production Counts": Production_count})
Production_df.to_csv("Resources/US_movies_Production.csv", encoding='utf-8', index=True)

Rated_df = pd.DataFrame({"Rated": Rated, "Rated Counts": Rated_count})
Rated_df.to_csv("Resources/US_movies_Rated.csv", encoding='utf-8', index=True)

Director_df = pd.DataFrame({"Director": Director, "Director Counts": Director_count})
Director_df.to_csv("Resources/US_movies_Director.csv", encoding='utf-8', index=True)
```

```python
Movies_df = pd.DataFrame({"Title": Title, 
                          "Runtime (mins)": Runtime,
                          "Year": Year,
                          "Release Date": Release_Date,
                          "Genre": Genres,
                          "IMDb Rating": Rating_imdb,
                          "Num Votes":Num_Votes,
                          "Rotten Tomatoes Rating": Rating_RT, 
                          "Metascore": Rating_Metacritic,
                          "Box Office": BoxOffice,
                          "Rated": Rates,
                          "Production":Productions,
                          "Director":Directors,
                          "Actor":Actors })

# Save to csv
Movies_df.to_csv("Resources/US_movies_cleanup.csv", encoding='utf-8', index=True)
```

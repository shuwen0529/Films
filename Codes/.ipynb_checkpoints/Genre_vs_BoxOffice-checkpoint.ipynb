```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

import locale
locale.setlocale( locale.LC_ALL, "en_US.UTF-8")

### Load file and get series
movie = pd.read_csv("../Resources/US_movies_cleanup.csv")
movie.drop(columns = ["Unnamed: 0"], inplace=True)
### Drop all NaN based on Box Office
movie.dropna(subset=["Box Office"], how='all', inplace = True)

Genre = movie["Genre"].tolist()        
BoxOffice = movie["Box Office"].tolist() 
L = len(BoxOffice)
```

```python
### Load Genre counts and get top10
movie_genre = pd.read_csv("../Resources/US_movies_Genre.csv")
movie_genre.sort_values(by=["Genre Counts"], ascending=False, inplace= True)
genre_top10 = movie_genre[:10].copy()["Genre"].tolist()
```

```python
### Get BoxOffice by Genre
BoxOffice_by_Genre = np.zeros(10)
G = len(genre_top10)
for g in range(G):
    genre = genre_top10[g]
    print(genre)
    for l in range(L):
        Genre_type  = [x.strip() for x in Genre[l].split(',')]
        if genre in Genre_type:
            BoxOffice_num = locale.atof(BoxOffice[l].strip("$"))
            BoxOffice_by_Genre[g] = BoxOffice_by_Genre[g]+BoxOffice_num
            
BoxOffice_billion =  BoxOffice_by_Genre/1e9
```

```python
### Plot Figure
fig, ax = plt.subplots(figsize=(10, 8))

plt.bar(range(G), BoxOffice_billion, alpha=0.7, edgecolor="k", lw=2)

rects = ax.patches
for rect, label in zip(rects, BoxOffice_billion):
    height = rect.get_height()
    ax.text(rect.get_x()+rect.get_width()/2, height, 
            int(np.round(label)),
            ha='center', va='bottom', fontsize=15, weight="bold")
    
plt.xticks(range(G), genre_top10, fontsize=15, weight="bold", rotation=60)
plt.yticks(np.arange(0,70,10), fontsize=15, weight="bold")
plt.ylabel("Box Office (Billion)", fontsize=15, weight="bold")

ax.xaxis.set_tick_params(width=3, size=10)
ax.yaxis.set_tick_params(width=3, size=10)

plt.tight_layout()
plt.savefig("../Figures/Movie_Genre_BoxOffice_BarPlot.png")
plt.show()
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

from matplotlib import pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
movie = pd.read_csv("Resources/US_movies_cleanup.csv")
movie.drop(columns = ["Unnamed: 0"], inplace=True)
movie.head(3)
```

```python
IMDb_rating = movie.dropna()["IMDb Rating"]*10
Metascore = movie.dropna()["Metascore"]
RT_rating = movie.dropna()["Rotten Tomatoes Rating"].str.rstrip('%').astype('float')

print(stats.ttest_ind(IMDb_rating, RT_rating))
print(stats.ttest_ind(IMDb_rating, Metascore))
print(stats.ttest_ind(RT_rating, Metascore))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

plt.hist(IMDb_rating, 30, density=True, facecolor='b', alpha=0.4, label="IMDb")
plt.hist(Metascore,   30, density=True, facecolor='g', alpha=0.4, label="Metascore")
n, bins, patches = plt.hist(RT_rating, 30, density=True, facecolor='orange', alpha=0.4, label="Rotten Tomatos")

# Add a 'best fit' line
IMDb_line = mlab.normpdf(bins, np.mean(IMDb_rating), np.std(IMDb_rating))
plt.plot(bins, IMDb_line, 'b--', linewidth =3)
Metascore_line = mlab.normpdf(bins, np.mean(Metascore), np.std(Metascore))
plt.plot(bins, Metascore_line, 'g--', linewidth =3)
RT_line = mlab.normpdf(bins, np.mean(RT_rating), np.std(RT_rating))
plt.plot(bins, RT_line, ls='--', c='orange', linewidth =3)

plt.legend(fontsize =16)
plt.xlabel('Rating', size=15, weight="bold")
plt.ylabel('Probability', size=15, weight="bold")
plt.tick_params(axis='both', which='major', labelsize=15)

plt.tight_layout()
plt.savefig("Figures/Movie_Rating_Distribution_histogram.png")

plt.show()
```
